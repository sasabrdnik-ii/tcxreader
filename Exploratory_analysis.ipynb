```python
import matplotlib.dates as mdates
from ipywidgets import interact, widgets
from IPython.display import display

# Ensure GOTR_TSTAKEN is in datetime format
df['GOTR_TSTAKEN'] = pd.to_datetime(df['GOTR_TSTAKEN'])

# Extract the year for filtering
df['Year'] = df['GOTR_TSTAKEN'].dt.year

# Filter the data for the year 2023
df_2023 = df[df['Year'] == 2023]

# Extract the day of year and company (OFFC_CODE) for grouping
df_2023['DayOfYear'] = df_2023['GOTR_TSTAKEN'].dt.dayofyear
df_2023['MonthDay'] = pd.to_datetime('2020-01-01') + pd.to_timedelta(df_2023['DayOfYear'] - 1, unit='days')

# Count the frequency of each date for each company (OFFC_CODE)
date_counts_2023 = df_2023.groupby(['OFFC_CODE', 'DayOfYear', 'MonthDay']).size().reset_index(name='Count')

# List of unique companies (OFFC_CODE)
unique_companies = df_2023['OFFC_CODE'].unique()

# Interactive function to plot based on selected company
def plot_office_data(selected_company):
    plt.figure(figsize=(16, 6))

    # Filter data for the selected company
    filtered_data = date_counts_2023[date_counts_2023['OFFC_CODE'] == selected_company]

    # Plot the filtered data
    sns.lineplot(x='MonthDay', y='Count', data=filtered_data, marker='o')

    # Format the x-axis to display month names
    plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%B'))

    # Add labels and title
    plt.xlabel('Month')
    plt.ylabel('Count')
    plt.title(f'Čas vnosa/prevzetja malice za leto 2023 - {selected_company}')

    plt.tight_layout()
    plt.show()

# Create an interactive dropdown widget
company_dropdown = widgets.Dropdown(
    options=unique_companies,
    value=unique_companies[0],  # Set default value
    description='OFFC_CODE:',
    disabled=False,
)

# Use ipywidgets interact to create interactivity
interact(plot_office_data, selected_company=company_dropdown)
```

```python
from scipy.stats import chi2_contingency
import numpy as np

df['GOTR_DATEDOCUMENT'] = pd.to_datetime(df['GOTR_DATEDOCUMENT'])

# Extract the day of the week from GOTR_DATEDOCUMENT
df['DayOfWeek'] = df['GOTR_DATEDOCUMENT'].dt.day_name()

# Create a contingency table
contingency_table = pd.crosstab(df['DayOfWeek'], df['GOTR_FLAG_TAKEN'])

# Calculate the Chi-squared statistic and p-value
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Calculate Cramér's V
n = contingency_table.sum().sum()  # total number of observations
phi2 = chi2 / n  # phi-square
r, k = contingency_table.shape  # number of rows and columns
cramers_v = np.sqrt(phi2 / min(k - 1, r - 1))

print("Korelacija med dnevom v tednu in (ne)prevzemom malice")
print("Cramér's V:", cramers_v)
print("Chi-squared Statistic:", chi2)
print("p-value:", p)
print("zelo nizka korelacija")
```

```python
def categorize_shift(shift_code):
    if '1' in shift_code:
        return 'Group 1'
    elif '2' in shift_code:
        return 'Group 2'
    elif '3' in shift_code:
        return 'Group 3'
    else:
        return 'Other'

# Apply the categorization function to the SHFT_CODE column
df['ShiftGroup'] = df['SHFT_CODE'].apply(categorize_shift)

# Create a contingency table based on ShiftGroup and GOTR_FLAG_TAKEN
contingency_table = pd.crosstab(df['ShiftGroup'], df['GOTR_FLAG_TAKEN'])

# Calculate the Chi-squared statistic and p-value
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Calculate Cramér's V
n = contingency_table.sum().sum()  # total number of observations
phi2 = chi2 / n  # phi-square
r, k = contingency_table.shape  # number of rows and columns
cramers_v = np.sqrt(phi2 / min(k - 1, r - 1))

# Print the results
print("Posplošitev izmen na 1-3")
print("Korelacija med kategorijo izmene (1-3) in prevzemom malice", cramers_v)
print("Chi-squared Statistic:", chi2)
print("p-value:", p)
print("korelacija je stat. pomembna, ampak izjemno nizka")
```
